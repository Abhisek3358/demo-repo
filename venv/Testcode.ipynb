```python
import numpy as np
# Create a sample array
arr = np.array([[1, 2, 3], [4, 5, 6]])

print("Array:")
print(arr)

# Compute sum of all elements
sum_arr = np.sum(arr)
print(f"\nSum of all elements: {sum_arr}")

# Compute mean of all elements
mean_arr = np.mean(arr)
print(f"Mean of all elements: {mean_arr}")
```
